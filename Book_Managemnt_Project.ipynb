```python
Book_Management = {
    "1": {"Title": "The C Programming", "Author": "F. Scott Fitzgerald"},
    "2": {"Title": "The Python", "Author": "Lee"},
    "3": {"Title": "The C++ Programming", "Author": "Orwell"},
    "4": {"Title": "The Great Gatsby", "Author": "F. Scott Fitzgerald"},
    "5": {"Title": "To Kill a Mockingbird", "Author": "Harper Lee"},
    "6": {"Title": "1984", "Author": "George Orwell"},
    "7": {"Title": "The Catcher in the Rye", "Author": "J.D. Salinger"},
    "8": {"Title": "Pride and Prejudice", "Author": "Jane Austen"}
}

def search_book(title):
    for book_id, book_info in Book_Management.items():
        if book_info["Title"].lower() == title.lower():
            return book_id
    return None


def add_book(title, author):
    new_id = str(len(Book_Management) + 1)
    Book_Management[new_id] = {"Title": title, "Author": author}
    

def remove_book(book_id):
    if book_id in Book_Management:
        del Book_Management[book_id]
        print("Book removed successfully.")
    else:
        print("Book is not Available")

def return_book(book_id):
    if book_id in Book_Management:
        print(f"Book details: {Book_Management[book_id]}")
    else:
        print("Book not found.")
        

while True:
    print("\nBook Management System")
    print("1. Search for a book")
    print("2. Add a book")
    print("3. Remove a book")
    print("4. Return a book")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        title = input("Enter the title of the book: ")
        book_id = search_book(title)
        if book_id:
            print(f"Book found: {Book_Management[book_id]}")
        else:
            print("Book not found.")

    elif choice == "2":
        title = input("Enter the title of the book: ")
        author = input("Enter the author name of the book: ")
        add_book(title, author)
        print("Book added successfully.")

    elif choice == "3":
        book_id = input("Enter the ID of the book: ")
        remove_book(book_id)

    elif choice == "4":
        book_id = input("Enter the ID of the book: ")
        return_book(book_id)

    elif choice == "5":
        print("Exiting")
        break

    else:
        print("Invalid choice")
```
